### Step 3 (EDA)

```python
data = data.drop(["ID", "MntWines"], axis=1)
```

```python
data["Dt_Customer"] = data["Dt_Customer"].astype(str).str[6:10].astype(int)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

data["Education"] = encoder.fit_transform(data["Education"])
data["Marital_Status"] = encoder.fit_transform(data["Marital_Status"])
```

```python
data["Income"] = data["Income"].fillna(data["Income"].mean())
```

### Step 4 (Make X and model)

```python
X = data.iloc[:, :]
```

```python
from sklearn.cluster import KMeans

model = KMeans()
```

### Step 5 (Ckeck which k is better )

```python
l = []
for i in range(1, 10):
    model = KMeans(n_clusters=i)
    model.fit(X)
    l.append(model.inertia_)
```

```python
plt.plot([1, 2, 3, 4, 5, 6, 7, 8, 9], l)
```

### Step 6 (Make new model and fit data)

```python
new_model = KMeans(init="k-means++", n_clusters=3, algorithm="lloyd")
```

```python
new_model.fit(X)
```

### Step 7 (Evaluation)

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(X, new_model.labels_)
print(score)
```
